# Teoría definicion de tablas en Django y tipos de campos

## Qué es un modelo ?


Un modelo es la fuente única y definitiva de información sobre tus datos. Contiene los campos y comportamientos esenciales de los datos que estás almacenando. Generalmente, cada modelo se corresponde con una única tabla de la base de datos.

Aspectos básicos:

Cada modelo es una clase de Python que hereda de django.db.models.Model.
Cada atributo del modelo representa un campo de la base de datos.
Con todo esto, Django te proporciona una API de acceso a la base de datos generada automáticamente; consulta Realizar consultas (Making queries).


## Ejemplo
from django.db import models


# class Person(models.Model):
#     first_name = models.CharField(max_length=30)
#     last_name = models.CharField(max_length=30)

first_name y last_name son campos del modelo. Cada campo se especifica como un atributo de clase, y cada atributo se mapea a una columna en la base de datos.
El modelo Person mencionado crearía una tabla en la base de datos similar a esta:

# CREATE TABLE myapp_person (
#     "id" bigint NOT NULL PRIMARY KEY GENERATED BY DEFAULT AS IDENTITY,
#     "first_name" varchar(30) NOT NULL,
#     "last_name" varchar(30) NOT NULL
# );
Algunas notas técnicas:

- El nombre de la tabla, myapp_person, se deriva automáticamente de algunos metadatos del modelo, pero puede ser sobrescrito. Consulta Nombres de         tablas para más detalles.

- Un campo id se agrega automáticamente como clave primaria, pero este comportamiento puede ser modificado. Consulta Campos de clave primaria             automática.

- El comando CREATE TABLE en este ejemplo está formateado utilizando la sintaxis de PostgreSQL, pero vale la pena señalar que Django genera SQL           adaptado al motor de base de datos especificado en tu archivo de configuración (settings.py).


Uso de modelos
Una vez que hayas definido tus modelos, necesitas indicarle a Django que vas a utilizarlos. Esto se hace editando el archivo de configuración (settings.py) y modificando la configuración INSTALLED_APPS para agregar el nombre del módulo que contiene tu archivo models.py.

Por ejemplo, si los modelos de tu aplicación están en el módulo myapp.models (la estructura de paquete que se crea al usar el comando manage.py startapp), INSTALLED_APPS debería incluir, en parte:

# INSTALLED_APPS = [
# ...
# "myapp",
# ...
# ]


Cuando agregues nuevas aplicaciones a INSTALLED_APPS, asegúrate de ejecutar manage.py migrate para aplicar los cambios en la base de datos. Opcionalmente, puedes hacer las migraciones primero con manage.py makemigrations para generar los archivos de migración.

El proceso es el siguiente:

makemigrations: Este comando genera los archivos de migración basados en los cambios realizados en los modelos de tus aplicaciones. Si agregas nuevos modelos o realizas cambios en los existentes, makemigrations crea un archivo de migración que Django usará para aplicar los cambios en la base de datos.

## Campo

La parte más importante de un modelo, y la única parte obligatoria, es la lista de campos de base de datos que define. Los campos se especifican mediante atributos de clase. Es importante tener cuidado de no elegir nombres de campos que entren en conflicto con la API del modelo, como clean, save o delete, ya que esos son métodos reservados en Django.

Algunos ejemplos de campos comunes en los modelos de Django son:

## CharField: Para cadenas de texto de longitud limitada.
## IntegerField: Para enteros.
## DateTimeField: Para fechas y horas.
## ForeignKey: Para definir una relación de clave externa con otro modelo.
## ManyToManyField: Para relaciones muchos a muchos entre modelos.

from django.db import models

Tipos de campos
Cada campo en tu modelo debe ser una instancia de la clase de campo apropiada. Django utiliza los tipos de clases de campo para determinar varias cosas:

- Tipo de columna: Le indica a la base de datos qué tipo de datos almacenar (por ejemplo, INTEGER, VARCHAR, TEXT).
- Widget HTML por defecto: El tipo de widget HTML que se debe usar cuando se renderiza un campo de formulario (por ejemplo, <input type="text">,          <select>).
- Requisitos mínimos de validación: Se usan en el panel de administración de Django y en los formularios generados automáticamente.
  Django incluye docenas de tipos de campo incorporados. A continuación, algunos de los tipos más comunes que puedes usar:

## Opciones de campo
Cada campo en un modelo de Django toma un conjunto específico de argumentos relacionados con ese campo, los cuales están documentados en la referencia de campos del modelo. Por ejemplo, el campo CharField (y sus subclases) requiere el argumento max_length, que especifica el tamaño del campo VARCHAR en la base de datos para almacenar los datos.

Además de los argumentos específicos de cada tipo de campo, hay un conjunto de argumentos comunes que están disponibles para todos los tipos de campo. Todos son opcionales y se explican en detalle en la referencia, pero aquí tienes un resumen rápido de los más utilizados:

1. null:

- Descripción: Si se establece en True, Django almacenará los valores vacíos como NULL en la base de datos.
  Valor por defecto: False (es decir, si no se especifica, los campos vacíos no se almacenan como NULL).
- Uso: Este parámetro está relacionado exclusivamente con la base de datos, ya que determina cómo se almacenan los valores vacíos en la base de datos.
  Ejemplo:
  email = models.EmailField(null=True)  #Almacena null
  En este caso, si no se proporciona un valor para email, se almacenará como NULL en la base de datos.

2. blank:

- Descripción: Si se establece en True, permite que el campo esté vacío en los formularios de validación.
- Valor por defecto: False (es decir, por defecto, los campos no se permiten vacíos en los formularios).
- Uso: Este parámetro está relacionado con la validación del formulario y determina si el campo es obligatorio en los formularios (como en el panel de    administración de Django). blank = True significa que dejango almacena una cadena vacía en la base de datos


" opciones (Choices)
Una secuencia de tuplas de 2 valores, un mapeo, un tipo de enumeración o una función callable (que no espera argumentos y devuelve cualquiera de los formatos anteriores), para usar como opciones para este campo. Si se proporciona esto, el widget de formulario predeterminado será un cuadro de selección en lugar del campo de texto estándar y limitará las opciones a las dadas.

Una lista de opciones tiene este aspecto:

# YEAR_IN_SCHOOL_CHOICES = [
#    ("FR", "Freshman"),
#    ("SO", "Sophomore"),
#    ("JR", "Junior"),
#    ("SR", "Senior"),
#    ("GR", "Graduate"),
# ]

El primer elemento de cada tupla es el valor que se almacenará en la base de datos. El segundo elemento es lo que se mostrará en el widget del formulario del campo.

Dada una instancia del modelo, el valor de visualización para un campo con opciones se puede acceder utilizando el método get_FOO_display(). Por ejemplo:

# Campos primarios automáticos
De forma predeterminada, Django asigna a cada modelo una clave primaria auto-incremental con el tipo especificado por aplicación en AppConfig.default_auto_field o de manera global en la configuración DEFAULT_AUTO_FIELD. Por ejemplo:

## id = models.BigAutoField(primary_key=True)


Si deseas especificar una clave primaria personalizada, utiliza primary_key=True en uno de tus campos. Si Django detecta que has configurado explícitamente Field.primary_key, no añadirá la columna automática id.

Cada modelo requiere exactamente un campo con primary_key=True (ya sea declarado explícitamente o añadido automáticamente).

## Relaciones
Claramente, el poder de las bases de datos relacionales radica en la capacidad de relacionar tablas entre sí. Django ofrece formas de definir los tres tipos más comunes de relaciones en bases de datos:

1. Relaciones muchos a uno (many-to-one).
2. Relaciones muchos a muchos (many-to-many).
3. Relaciones uno a uno (one-to-one).

### Ejemplo
### from django.db import models


### class Manufacturer(models.Model):
###    # ...
###    pass


### class Car(models.Model):
###    manufacturer = models.ForeignKey(Manufacturer, on_delete=models.CASCADE)
###    # ...


## Relaciones muchos a muchos (Many-to-many relationships)
Para definir una relación muchos a muchos, utiliza ManyToManyField. Se usa como cualquier otro tipo de campo, incluyéndolo como un atributo de clase en tu modelo.

ManyToManyField requiere un argumento posicional: la clase con la que el modelo está relacionado.

Por ejemplo, si una Pizza tiene múltiples objetos Topping (ingredientes), es decir, un Topping puede estar en múltiples pizzas y cada Pizza tiene múltiples ingredientes, aquí está cómo se representaría:

### from django.db import models


### class Topping(models.Model):
###    # ...
###    pass


### class Pizza(models.Model):
###    # ...
###    toppings = models.ManyToManyField(Topping)

Los campos ManyToManyField también aceptan una serie de argumentos adicionales que se explican en la referencia de campos del modelo. Estas opciones ayudan a definir cómo debe funcionar la relación; todas son opcionales.

Campos adicionales en relaciones muchos a muchos
Cuando solo se trata de relaciones muchos a muchos, como combinar y emparejar pizzas y ingredientes, un ManyToManyField estándar es todo lo que necesitas. Sin embargo, en ocasiones, puede ser necesario asociar datos con la relación entre dos modelos

Por ejemplo, considera el caso de una aplicación que rastrea los grupos musicales a los que pertenecen los músicos. Existe una relación muchos a muchos entre una persona y los grupos de los que es miembro, por lo que podrías usar un ManyToManyField para representar esta relación. Sin embargo, podrías querer recopilar detalles adicionales sobre la membresía, como la fecha en que la persona se unió al grupo.

Para estas situaciones, Django permite especificar el modelo que se usará para gestionar la relación muchos a muchos. Luego, puedes agregar campos adicionales al modelo intermedio. Este modelo intermedio se asocia con el ManyToManyField utilizando el argumento through para señalar el modelo que actuará como intermediario. Para nuestro ejemplo de músicos, el código se vería algo así:



El modelo Membership actúa como modelo intermedio entre Person y Group.
El campo date_joined en Membership permite registrar la fecha en que una persona se unió a un grupo.
El campo invite_reason proporciona información adicional sobre por qué se invitó a la persona al grupo.
Este enfoque te permite agregar cualquier cantidad de información adicional sobre la relación muchos a muchos.

la clase Membership se utiliza para "romper" la relación muchos a muchos entre las clases Person y Group al actuar como un modelo intermedio.

En un escenario típico de relación muchos a muchos, Django manejaría automáticamente la creación de una tabla intermedia para gestionar las relaciones entre los modelos, sin necesidad de que tú lo definas explícitamente. Sin embargo, cuando necesitas asociar más datos a esa relación (como en este caso date_joined o invite_reason), puedes definir un modelo intermedio, como lo has hecho con Membership.

# Managers en Django

En django los managers son una forma de encapsular la lógica de consulta de base de datos en los modelos. Son objetos que permiten interactuar con las tablas de la base de datos de una manera mas flexible y ordenada.
Por defecto, cada modelo de Django tiene un manager llamado objects, que proporciona métodos básicos como ,all(), filter(), exclude(), get(), etc, para realizar consultas a la base de datos. Sin embargo, si queremos agregar métodos personalizados a estas consultas, o modificar la forma en que django realiza las consultas, podemos crear nuestro propio manager.

# Como funcionan los managers?
1. Definir un manager personalizado: Podemos crear una clase que herede a models.Manager y luego       agregarle nuestros propios métodos de consulta.
2. Usar el manager en el modelo: Despues de definir un manager personalizado, hay que agregarlo        como un atributo en la clase del modelo, lo que permite usarlo para consultas especializadas.

## Filtros con exclusión Exclude

🔍 Filtros con exclusión en Django usando exclude()
En Django, el método exclude() se usa para filtrar registros excluyendo aquellos que coincidan con una condición específica. Es el opuesto de filter(), que selecciona solo los registros que cumplen con la condición dada.


# Listado de lookups (operadores de comparación) en Django para consultas (funcionan en filter() y exclude()):

1. ## Comparación de igualdad:
   - exact: Campo igual a un valor.
     Ejemplo: .exclude(campo__exact=valor) o simplemente .exclude(campo=valor)
   - iexact: Igual que exact, pero sin distinguir mayúsculas y minúsculas.
     Ejemplo: .exclude(campo__iexact=valor)

2. ## Búsquedas por subcadena:
   - contains: Comprueba que el campo contenga la subcadena (sensible a mayúsculas).
     Ejemplo: .exclude(campo__contains="texto")
   - icontains: Igual que contains, pero sin distinguir mayúsculas/minúsculas.
     Ejemplo: .exclude(campo__icontains="texto")

3. ## Comparación de pertenencia:
   - in: Verifica que el valor del campo esté en un iterable (lista, tupla, etc.).
     Ejemplo: .exclude(campo__in=[valor1, valor2, valor3])

4. ## Comparaciones numéricas:
   - gt: Mayor que.
     Ejemplo: .exclude(campo__gt=valor)
   - gte: Mayor o igual que.
     Ejemplo: .exclude(campo__gte=valor)
   - lt: Menor que.
     Ejemplo: .exclude(campo__lt=valor)
   - lte: Menor o igual que.
     Ejemplo: .exclude(campo__lte=valor)

5. ## Búsquedas en cadenas:
   - startswith: Comprueba si el campo empieza con un determinado valor (sensible a mayúsculas).
     Ejemplo: .exclude(campo__startswith="prefijo")
   - istartswith: Igual que startswith, pero sin distinguir mayúsculas/minúsculas.
     Ejemplo: .exclude(campo__istartswith="prefijo")
   - endswith: Comprueba si el campo termina con un determinado valor (sensible a mayúsculas).
     Ejemplo: .exclude(campo__endswith="sufijo")
   - iendswith: Igual que endswith, pero sin distinguir mayúsculas/minúsculas.
     Ejemplo: .exclude(campo__iendswith="sufijo")

6. ## Rango de valores:
   - range: Comprueba que el valor del campo se encuentre dentro de un rango (incluye ambos extremos).
     Ejemplo: .exclude(campo__range=(valor_min, valor_max))

7. ## Valores nulos:
   - isnull: Comprueba si el campo es NULL o no.
     Ejemplo: .exclude(campo__isnull=True)  // Excluye donde el campo es nulo

8. ## Búsqueda mediante expresiones regulares:
   - regex: Filtra usando una expresión regular (sensible a mayúsculas).
     Ejemplo: .exclude(campo__regex=r'expresion_regular')
   - iregex: Igual que regex, pero sin distinguir mayúsculas.
     Ejemplo: .exclude(campo__iregex=r'expresion_regular')

9. ## Lookups específicos para campos de fecha y hora:
   - year: Extrae y compara el año de un campo de fecha.
     Ejemplo: .exclude(campo_fecha__year=2025)
   - month: Extrae y compara el mes.
     Ejemplo: .exclude(campo_fecha__month=2)
   - day: Extrae y compara el día.
     Ejemplo: .exclude(campo_fecha__day=15)
   - week_day: Filtra por el día de la semana (según la configuración de la base de datos).
     Ejemplo: .exclude(campo_fecha__week_day=2)

Estos operadores son aplicables tanto a consultas con filter() como con exclude(), permitiéndote construir condiciones de inclusión o exclusión según necesites.

from django.db import models
from django.db.models import Q

class AutorManager(models.Manager):
    def buscar_autor_por_nombre(self, kword):
        return self.filter(nombre__icontains=kword)
    
    def excluir_autores_por_edad(self, edad):
        return self.exclude(edad=edad)
    
    def buscar_y_excluir(self, kword, edad_excluir):
        # Busca autores cuyo nombre contenga la palabra clave y excluye los de edad específica
        return self.filter(nombre__icontains=kword).exclude(edad=edad_excluir)

class Autor(models.Model):
    nombre = models.CharField(max_length=50)
    apellidos = models.CharField(max_length=50)
    edad = models.PositiveIntegerField()
    
    objects = AutorManager()

    def __str__(self):
        return f"{self.nombre} {self.apellidos}
