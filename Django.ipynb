## Qu es un modelo ?


Un modelo es la fuente nica y definitiva de informacin sobre tus datos. Contiene los campos y comportamientos esenciales de los datos que ests almacenando. Generalmente, cada modelo se corresponde con una nica tabla de la base de datos.

Aspectos bsicos:

Cada modelo es una clase de Python que hereda de django.db.models.Model.
Cada atributo del modelo representa un campo de la base de datos.
Con todo esto, Django te proporciona una API de acceso a la base de datos generada automticamente; consulta Realizar consultas (Making queries).


## Ejemplo
from django.db import models


# class Person(models.Model):
#     first_name = models.CharField(max_length=30)
#     last_name = models.CharField(max_length=30)

first_name y last_name son campos del modelo. Cada campo se especifica como un atributo de clase, y cada atributo se mapea a una columna en la base de datos.
El modelo Person mencionado creara una tabla en la base de datos similar a esta:

# CREATE TABLE myapp_person (
#     "id" bigint NOT NULL PRIMARY KEY GENERATED BY DEFAULT AS IDENTITY,
#     "first_name" varchar(30) NOT NULL,
#     "last_name" varchar(30) NOT NULL
# );
Algunas notas tcnicas:

- El nombre de la tabla, myapp_person, se deriva automticamente de algunos metadatos del modelo, pero puede ser sobrescrito. Consulta Nombres de         tablas para ms detalles.

- Un campo id se agrega automticamente como clave primaria, pero este comportamiento puede ser modificado. Consulta Campos de clave primaria             automtica.

- El comando CREATE TABLE en este ejemplo est formateado utilizando la sintaxis de PostgreSQL, pero vale la pena sealar que Django genera SQL           adaptado al motor de base de datos especificado en tu archivo de configuracin (settings.py).

## Opciones de campo
Cada campo en un modelo de Django toma un conjunto especfico de argumentos relacionados con ese campo, los cuales estn documentados en la referencia de campos del modelo. Por ejemplo, el campo CharField (y sus subclases) requiere el argumento max_length, que especifica el tamao del campo VARCHAR en la base de datos para almacenar los datos.

Adems de los argumentos especficos de cada tipo de campo, hay un conjunto de argumentos comunes que estn disponibles para todos los tipos de campo. Todos son opcionales y se explican en detalle en la referencia, pero aqu tienes un resumen rpido de los ms utilizados:

1. null:

- Descripcin: Si se establece en True, Django almacenar los valores vacos como NULL en la base de datos.
  Valor por defecto: False (es decir, si no se especifica, los campos vacos no se almacenan como NULL).
- Uso: Este parmetro est relacionado exclusivamente con la base de datos, ya que determina cmo se almacenan los valores vacos en la base de datos.
  Ejemplo:
  email = models.EmailField(null=True)  #Almacena null
  En este caso, si no se proporciona un valor para email, se almacenar como NULL en la base de datos.

2. blank:

- Descripcin: Si se establece en True, permite que el campo est vaco en los formularios de validacin.
- Valor por defecto: False (es decir, por defecto, los campos no se permiten vacos en los formularios).
- Uso: Este parmetro est relacionado con la validacin del formulario y determina si el campo es obligatorio en los formularios (como en el panel de    administracin de Django). blank = True significa que dejango almacena una cadena vaca en la base de datos

## Relaciones muchos a muchos (Many-to-many relationships)
Para definir una relacin muchos a muchos, utiliza ManyToManyField. Se usa como cualquier otro tipo de campo, incluyndolo como un atributo de clase en tu modelo.

ManyToManyField requiere un argumento posicional: la clase con la que el modelo est relacionado.

Por ejemplo, si una Pizza tiene mltiples objetos Topping (ingredientes), es decir, un Topping puede estar en mltiples pizzas y cada Pizza tiene mltiples ingredientes, aqu est cmo se representara:

### from django.db import models


### class Topping(models.Model):
###    # ...
###    pass


### class Pizza(models.Model):
###    # ...
###    toppings = models.ManyToManyField(Topping)

Los campos ManyToManyField tambin aceptan una serie de argumentos adicionales que se explican en la referencia de campos del modelo. Estas opciones ayudan a definir cmo debe funcionar la relacin; todas son opcionales.

Campos adicionales en relaciones muchos a muchos
Cuando solo se trata de relaciones muchos a muchos, como combinar y emparejar pizzas y ingredientes, un ManyToManyField estndar es todo lo que necesitas. Sin embargo, en ocasiones, puede ser necesario asociar datos con la relacin entre dos modelos

Por ejemplo, considera el caso de una aplicacin que rastrea los grupos musicales a los que pertenecen los msicos. Existe una relacin muchos a muchos entre una persona y los grupos de los que es miembro, por lo que podras usar un ManyToManyField para representar esta relacin. Sin embargo, podras querer recopilar detalles adicionales sobre la membresa, como la fecha en que la persona se uni al grupo.

Para estas situaciones, Django permite especificar el modelo que se usar para gestionar la relacin muchos a muchos. Luego, puedes agregar campos adicionales al modelo intermedio. Este modelo intermedio se asocia con el ManyToManyField utilizando el argumento through para sealar el modelo que actuar como intermediario. Para nuestro ejemplo de msicos, el cdigo se vera algo as:



El modelo Membership acta como modelo intermedio entre Person y Group.
El campo date_joined en Membership permite registrar la fecha en que una persona se uni a un grupo.
El campo invite_reason proporciona informacin adicional sobre por qu se invit a la persona al grupo.
Este enfoque te permite agregar cualquier cantidad de informacin adicional sobre la relacin muchos a muchos.

la clase Membership se utiliza para "romper" la relacin muchos a muchos entre las clases Person y Group al actuar como un modelo intermedio.

En un escenario tpico de relacin muchos a muchos, Django manejara automticamente la creacin de una tabla intermedia para gestionar las relaciones entre los modelos, sin necesidad de que t lo definas explcitamente. Sin embargo, cuando necesitas asociar ms datos a esa relacin (como en este caso date_joined o invite_reason), puedes definir un modelo intermedio, como lo has hecho con Membership.

# Managers en Django

En django los managers son una forma de encapsular la lgica de consulta de base de datos en los modelos. Son objetos que permiten interactuar con las tablas de la base de datos de una manera mas flexible y ordenada.
Por defecto, cada modelo de Django tiene un manager llamado objects, que proporciona mtodos bsicos como ,all(), filter(), exclude(), get(), etc, para realizar consultas a la base de datos. Sin embargo, si queremos agregar mtodos personalizados a estas consultas, o modificar la forma en que django realiza las consultas, podemos crear nuestro propio manager.

# Como funcionan los managers?
1. Definir un manager personalizado: Podemos crear una clase que herede a models.Manager y luego       agregarle nuestros propios mtodos de consulta.
2. Usar el manager en el modelo: Despues de definir un manager personalizado, hay que agregarlo        como un atributo en la clase del modelo, lo que permite usarlo para consultas especializadas.

 Filtros con exclusin en Django usando exclude()
En Django, el mtodo exclude() se usa para filtrar registros excluyendo aquellos que coincidan con una condicin especfica. Es el opuesto de filter(), que selecciona solo los registros que cumplen con la condicin dada.


# Listado de lookups (operadores de comparacin) en Django para consultas (funcionan en filter() y exclude()):

1. ## Comparacin de igualdad:
   - exact: Campo igual a un valor.
     Ejemplo: .exclude(campo__exact=valor) o simplemente .exclude(campo=valor)
   - iexact: Igual que exact, pero sin distinguir maysculas y minsculas.
     Ejemplo: .exclude(campo__iexact=valor)

2. ## Bsquedas por subcadena:
   - contains: Comprueba que el campo contenga la subcadena (sensible a maysculas).
     Ejemplo: .exclude(campo__contains="texto")
   - icontains: Igual que contains, pero sin distinguir maysculas/minsculas.
     Ejemplo: .exclude(campo__icontains="texto")

3. ## Comparacin de pertenencia:
   - in: Verifica que el valor del campo est en un iterable (lista, tupla, etc.).
     Ejemplo: .exclude(campo__in=[valor1, valor2, valor3])

4. ## Comparaciones numricas:
   - gt: Mayor que.
     Ejemplo: .exclude(campo__gt=valor)
   - gte: Mayor o igual que.
     Ejemplo: .exclude(campo__gte=valor)
   - lt: Menor que.
     Ejemplo: .exclude(campo__lt=valor)
   - lte: Menor o igual que.
     Ejemplo: .exclude(campo__lte=valor)

5. ## Bsquedas en cadenas:
   - startswith: Comprueba si el campo empieza con un determinado valor (sensible a maysculas).
     Ejemplo: .exclude(campo__startswith="prefijo")
   - istartswith: Igual que startswith, pero sin distinguir maysculas/minsculas.
     Ejemplo: .exclude(campo__istartswith="prefijo")
   - endswith: Comprueba si el campo termina con un determinado valor (sensible a maysculas).
     Ejemplo: .exclude(campo__endswith="sufijo")
   - iendswith: Igual que endswith, pero sin distinguir maysculas/minsculas.
     Ejemplo: .exclude(campo__iendswith="sufijo")

6. ## Rango de valores:
   - range: Comprueba que el valor del campo se encuentre dentro de un rango (incluye ambos extremos).
     Ejemplo: .exclude(campo__range=(valor_min, valor_max))

7. ## Valores nulos:
   - isnull: Comprueba si el campo es NULL o no.
     Ejemplo: .exclude(campo__isnull=True)  // Excluye donde el campo es nulo

8. ## Bsqueda mediante expresiones regulares:
   - regex: Filtra usando una expresin regular (sensible a maysculas).
     Ejemplo: .exclude(campo__regex=r'expresion_regular')
   - iregex: Igual que regex, pero sin distinguir maysculas.
     Ejemplo: .exclude(campo__iregex=r'expresion_regular')

9. ## Lookups especficos para campos de fecha y hora:
   - year: Extrae y compara el ao de un campo de fecha.
     Ejemplo: .exclude(campo_fecha__year=2025)
   - month: Extrae y compara el mes.
     Ejemplo: .exclude(campo_fecha__month=2)
   - day: Extrae y compara el da.
     Ejemplo: .exclude(campo_fecha__day=15)
   - week_day: Filtra por el da de la semana (segn la configuracin de la base de datos).
     Ejemplo: .exclude(campo_fecha__week_day=2)

Estos operadores son aplicables tanto a consultas con filter() como con exclude(), permitindote construir condiciones de inclusin o exclusin segn necesites.

from django.db import models
from django.db.models import Q

class AutorManager(models.Manager):
    def buscar_autor_por_nombre(self, kword):
        return self.filter(nombre__icontains=kword)
    
    def excluir_autores_por_edad(self, edad):
        return self.exclude(edad=edad)
    
    def buscar_y_excluir(self, kword, edad_excluir):
        # Busca autores cuyo nombre contenga la palabra clave y excluye los de edad especfica
        return self.filter(nombre__icontains=kword).exclude(edad=edad_excluir)

class Autor(models.Model):
    nombre = models.CharField(max_length=50)
    apellidos = models.CharField(max_length=50)
    edad = models.PositiveIntegerField()
    
    objects = AutorManager()

    def __str__(self):
